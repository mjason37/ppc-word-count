```python
# Import required modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from nltk.corpus import stopwords
stop = set(stopwords.words("english"))
```

```python
# Read highspeend training and virtual college ppc data
hp = pd.read_excel('virtual-college.co.uk-adwords.Positions-uk-20191111-2019-11-12T06_05_40Z.xlsx')
vc = pd.read_excel('highspeedtraining.co.uk-adwords.Positions-uk-20191111-2019-11-12T06_04_51Z.xlsx')
```

```python
# Take a look at the 1st df
hp.head()
```

```python
# Create a function to process and clean the text data

def processVariable(data):
    """data = df to process"""
    
    # Remove nans to avoid dtypes error
    data.dropna(subset=["Title", "Description"],inplace=True)
    
    # Calculate description character length
    data["desCharLen"] = data["Description"].str.len()
    
    # Drop rows where length of description is greater the 180 character
#     to_drop = data.loc[data.des_char_len>180].index
#     data.drop(to_drop,axis=0,inplace=True)
#     data = data.reset_index(drop=True)
    
    data = data[data.desCharLen<=180].copy(deep=False)
    
    # Remove |, & from title and desciption
    data["Title"] = data["Title"].str.replace("|","").str.replace("&","")
    data["Description"] = data["Description"].str.replace("&","")
    
    # Now remove stop words from title and decription
    data["Title"] = data["Title"].apply(lambda x: " ".join([word for word in x.split() if word not in (stop)])).str.lower()
    data["Description"] = data["Description"].apply(lambda x: " ".join([word for word in x.split() if word not in (stop)])).str.lower()
    return data




# Function to display bar labels in bar plot
def absBarLabels():
    font_size = 15
    plt.ylabel("Words Count", fontsize = font_size)
    plt.xticks(rotation = 90, fontsize = font_size)
    plt.yticks([])
    
    # Set individual bar lebels in absolute number
    for x in ax.patches:
        ax.annotate(x.get_height(), 
        (x.get_x() + x.get_width()/2., x.get_height()), ha = "center", va = "center", xytext = (0, 7), 
        textcoords = "offset points", fontsize = font_size, color = "red")
        
        
def absoluteFreq(variable, title):
    global  ax
    
    # Dataframe of absolute and relative frequency
    absoluteFreq = variable.value_counts().head(30)
    df = pd.DataFrame({"Absolute Frequency":absoluteFreq})
    # This portion plots absolute frequency with bar labeled.
    fig_size = (18,8)
    font_size = 15
    title_size = 18
    ax =  absoluteFreq.plot.bar(title = title, figsize = fig_size, color ="teal")
    ax.title.set_size(title_size)
    absBarLabels()  # Displays bar labels in abs scale.
    sns.despine()
```

```python
# Fequency of words in highspeed training ppc title
hp = processVariable(hp)
absoluteFreq(hp.Title.str.split(expand=True).stack(),"Fequency of Words (Top 30) Used in Highspeed Training PPC Title")
```

```python
# Fequency of words in virtual college ppc title
vc = processVariable(vc)
absoluteFreq(vc.Title.str.split(expand=True).stack(),"Fequency of Words (Top 30) Used in Virtual College PPC Title")
```

```python
# Fequency of words in highspeed training ppc description
absoluteFreq(hp.Description.str.split(expand=True).stack(),"Fequency of Words (Top 30) Used in Highspeed Training PPC Description")
```

```python
# Fequency of words in virtual college ppc description
absoluteFreq(vc.Description.str.split(expand=True).stack(),"Fequency of Words (Top 30) Used in Virtual College PPC Description")
```

```python
# We can visualize those info using word cloud
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)

def plotWorldCloud(data, title):
    wordcloud = WordCloud(
        background_color="white",
        stopwords=stopwords,
        max_words=250,
        max_font_size=40, 
        scale=3,
        random_state=1 
    ).generate(str(data))

    fig = plt.figure(1, figsize=(25, 12))
    plt.axis("off")
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.title(title,fontsize=25)
    plt.show()
```

```python
# Word clouds of highspeed training ppc title
plotWorldCloud(hp.Title, "Title: Word Cloud of Words Used in Highspeed Training PPC Title")
```

```python
# Word clouds of virtual college ppc title
plotWorldCloud(vc.Title, "Title: Word Cloud of Words Used in Virtual College PPC Title")
```

```python
# Word clouds of highspeed training ppc description
plotWorldCloud(hp.Description, "Title: Word Cloud of Words Used in Highspeed Training PPC Description")
```

```python
# Word clouds of virtual college ppc description
plotWorldCloud(vc.Description, "Title: Word Cloud of Words Used in Virtual College PPC Description")
```
